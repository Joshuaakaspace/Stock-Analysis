```python
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
```

```python
IQR = q3 - q1
```

```python
IQR
```

```python
sns.boxplot(df['alk_phosphate'])
```

```python
(df < (q1 -1.5*IQR))| (df>(q3+1.5*IQR))
```

```python
df_no_outlier = df[~((df < (q1 -1.5*IQR))| (df>(q3+1.5*IQR))).any(axis=1)]
```

```python
df_no_outlier
```
